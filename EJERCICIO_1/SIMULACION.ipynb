```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### FUNCION SIMULACION DE LANZAMIENTO

```python
def simular_lanzamientos(p, N=1000):
    resultados = []
    for i in range(N):
        lanzamientos = 0
        while np.random.rand() > p:
            lanzamientos += 1
        resultados.append(lanzamientos + 1)
    return resultados
```

### FUNCION GRAFICAS CON SEABORN

```python
def graficar_distribucion(resultados, p):
    sns.histplot(resultados, bins=range(1, max(resultados) + 2), kde=False, stat="density")
    plt.xlabel("Número de lanzamientos hasta el primer éxito")
    plt.ylabel("Densidad")
    plt.title(f"Distribución de lanzamientos para p={p}")
    plt.show()
```

### ASIGNACION DE VALORES Y EJECUCION

```python
# Simulación y visualización con distintos valores de p
valores_p = [0.2, 0.5, 0.7]
N = 1000

for p in valores_p:
    resultados = simular_lanzamientos(p, N)
    graficar_distribucion(resultados, p)
```
